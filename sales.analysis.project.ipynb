```python
#Task No 2: Fill the empty boxes of column 'Price'
#Find the average (mean) price of all the items, and put that average number into the empty price boxes.
average_value = df['Price'].mean()
df['Price'] = df['Price'].fillna(average_value)
```

```python
df['Price']
```

```python
#Task 3: Fix the Minus Numbers
#In the 'Quantity' column, some numbers are negative (like -2 or -3). You cannot buy -2 items. Change these negative numbers into normal, positive numbers (like 2 or 3).
# Convert negative quantities to positive
df['Quantity'] = df['Quantity'].abs()
```

```python
average_value2 = df['Quantity'].mean()
df['Quantity'] = df['Quantity'].fillna(average_value2)
```

```python
df['Quantity']
```

```python
#Task 4: Calculate the Total
#Create a brand new column and name it 'Total_Sales'.
df['Total_Sales'] = df['Price'] * df['Quantity']
```

```python
df['Total_Sales']
```

```python
#Task 5: Find the Best City
#Now that you have the total sales, group the data by 'City'. Which city (Karachi, Lahore, or Islamabad) made the most money overall?
City_Totals = df.groupby('City')['Total_Sales'].sum()
best_City = City_Totals.nlargest(1)
print(best_City)
```
